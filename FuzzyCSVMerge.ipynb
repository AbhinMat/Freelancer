```python
import pandas as pd
import fuzzy_pandas as fpd

#read data into 2 pandas dataframe
df1 = pd.read_csv("table1.csv")
df2 = pd.read_csv("table2.csv")

#since CUSTOMER_NAME includes buyer/seller, it would be better to segregate it and create a new
#dataframe df3 with separated names
df3 = pd.DataFrame(df1.CUSTOMER_NAME.str.split('(',1).tolist(),
                                 columns = ['CUSTOMER_NAMES','BUYERORSELLER'])
#combining df1 and df3 into one dataframe df4
df4 = pd.concat([df1, df3], axis=1, sort=False)

#using merge function of fuzzy function to find copies
matches = fpd.fuzzy_merge(df4, df2,
                          left_on=['CUSTOMER_NAMES'],
                          right_on=['NAME'],
                          keep_left = df1.columns.tolist(),
                          keep_right = df2.columns.tolist(),
                          ignore_case=True,
                          ignore_order_words=True,
                          method = 'levenshtein',
                          threshold = 0.74,
                                
                          )

#since table2.csv contains some unformatted data which will create NaN values
#matches.dropna() is used to drop NaN values
matches=matches.dropna()
print(matches.dropna())

#writing resulting dataframe into 
matches.to_csv('final', index=False)
```
